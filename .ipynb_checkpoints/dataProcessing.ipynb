```python
import numpy as np
import pandas as pd
import matplotlib as plt

dataset = pd.read_csv('/home/tansen/my_files/self Learning/machineLearningFromUdemyA2Z/Data.csv')
# take all the column except the last ones 2 mean all except last 2 and : mean simply all
# x for feature mean indepent variable and y for depended variable
x = dataset.iloc[:,:-1].values
# takes the values of last columns
y = dataset.iloc[:,-1].values
print("values of X")
print(x)
print("values of Y")
print(y)

```

```python
# handling the missing values
from sklearn.impute import SimpleImputer
# creating the imputer object that will transform numberical missing values with mean for 1st and 2nd column only it have to be numerical
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(x[: , 1 : 3])
x[: , 1 : 3] = imputer.transform(x[: , 1 : 3])
print(x)
```

```python
# transform categorial values to to numercal for train the model
# this is for indepent variable 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# for 1st column it transform the countries name to categorial values 
ct = ColumnTransformer(transformers = [('encoder' , OneHotEncoder(),[0])] , remainder = 'passthrough' )
x = np.array(ct.fit_transform(x))
print(x)
```

```python
# for dependent variable which we are trying to predict
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
# splict test and train set for our data
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(x , y , test_size = .2 , random_state = 1)
print("X train" , X_train)
print("X test ", X_test)
print("Y train" , Y_train)
print("Y test ", Y_test)
```
